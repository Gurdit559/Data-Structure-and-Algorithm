## Implementing Singly Linked list

```python
#linked list

class Node():
    def __init__(self,value):
        self.value=value
        self.next=None
```

```python
class LinkedList():

    def __init__ (self):
        self.head = None #not string,not integer. only placeholder
        self.tail = None
        self.length = 0

    def append(self, value):
        #create a container node to hold value
        new_node = Node(value)
        #if its first node of list assign head and tail
        if self.length == 0:
            self.head = new_node
            self.tail = new_node
            self.length += 1
            return

        #add node to tail
        self.tail.next = new_node
        #modify tail to latest node
        self.tail = new_node
        #increment length
        self.length +=1

    def prepend(self, value):
        #create a container node to hold value
        new_node = Node(value)
        new_node.next = self.head
        self.head = new_node
        self.length += 1

    def insert(self, value,index):
        #check params --> index inbound
        #create a container node to hold value
        new_node = Node(value)
        temp_head = self.head
        #run for index times
        for i in range(index-1):
            temp_head = temp_head.next

        leader = temp_head
        follower = leader.next
        leader.next = new_node
        new_node.next = follower
        self.length += 1

    def remove(self,index):
        if index>=0 and index<=self.length-1:
            # = self.length - 1
            #if index given-remove the given Node
            #otherwise delete last node
            if index==0: #modify head
                self.head=self.head.next
                self.length -= 1

            elif index == self.length-1: #modify tail
                temp_head = self.head

                for i in range(index-2):
                    temp_head = temp_head.next

                self.tail=temp_head.next
                self.tail.next=None
                self.length -= 1

            else: #(majority-cases)
                temp_head = self.head
                for i in range(index-1):
                    temp_head = temp_head.next

                leader = temp_head
                follower = leader.next
                leader.next = follower.next
                self.length -= 1
        else:
            print('enter valid index')
            
    def reverse(self):
        prev=None
        h=self.head
        while h:
            temp=h
            h=h.next
            temp.next=prev
            prev=temp
        return prev

    def val_at_ind(self, index):
        temp_head = self.head
        #run for index times
        for i in range(index):
            temp_head = temp_head.next
        return temp_head.value

    def find(self, data):
        #return true if found else false
        temp_head = self.head
        while(temp_head != None):
            if temp_head.value == data:
                return True
            temp_head = temp_head.next #i=+1
            return False

    def printLL(self):
        temp_head = self.head
        while(temp_head):   #object at  = True      None = False
            print(temp_head.value)
            temp_head = temp_head.next
```

```python
myLL = LinkedList()
myLL.append(10)
myLL.append(20)
myLL.append(30)
myLL.printLL()

print("------------------")
myLL.prepend(7)
myLL.append(50)
myLL.prepend(3)

myLL.insert(4002, 2)
myLL.printLL()
print("\n")
myLL.remove(1)

print("------------------")
myLL.printLL()
```
